This is a code which illustrates the electric field distribution of similarly charged point charges. The charges are 1 and 1 Coloumb respectively. They are located at (1,1) & (-1,-1).

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
q1 = 1e-6  # Charge magnitude in Coulombs
k = 1 / (4 * np.pi * 8.85e-12)  # Coulomb's constant

# Charge positions
charge_positions = [(-1, 0), (1, 0)]

# Create a grid of points in the x-y plane
x_grid, y_grid = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))

# Initialize electric field components
Ex = np.zeros(x_grid.shape)
Ey = np.zeros(y_grid.shape)

# Calculate the electric field due to each charge
for (x1, y1) in charge_positions:
    # Calculate the distance squared from each point in the grid to the charge
    r_squared = (x_grid - x1)**2 + (y_grid - y1)**2
    r = np.sqrt(r_squared)

    # Avoid division by zero by setting r to a small value where r is zero
    r[r == 0] = 1e-20

    # Calculate the electric field components
    Ex += k * q1 * (x_grid - x1) / r_squared**1.5
    Ey += k * q1 * (y_grid - y1) / r_squared**1.5

# Plotting the electric field
plt.figure(figsize=(10, 10))
plt.streamplot(x_grid, y_grid, Ex, Ey, color='blue')
plt.scatter(*zip(*charge_positions), color='yellow', s=100, label='Charge Locations')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Electric Field of Two Equal Positive Charges')
plt.axhline(0, color='grey', lw=0.5, ls='--')  # Add x-axis
plt.axvline(0, color='grey', lw=0.5, ls='--')  # Add y-axis
plt.xlim(-4, 4)
plt.ylim(-4, 4)
plt.grid()
plt.legend()
plt.show()
```

This is a code which illustrates the magnetic field distribution when a current carrying loop is placed perpendicular to a planar surface. Here the loop intersects XY Plane at (-2,0) and (2,0) and is kept upright in Z direction.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points to evaluate the electric field
x_grid, y_grid = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))

dtheta = (2*np.pi)/100
R = 2
I = 5
k = 10**(-7)*R*I

Bx, By = 0, 0

for theta in np.linspace(0, 2*np.pi, 100):
    Bx += (-y_grid*np.cos(theta)*k*dtheta)/((x_grid**2 + y_grid**2 + R**2 + 2*x_grid*R*np.cos(theta))**(1.5))
    By += ((x_grid*np.cos(theta)+R)*k*dtheta)/((x_grid**2 + y_grid**2 + R**2 + 2*x_grid*R*np.cos(theta))**(1.5))

# Plot the electric field lines
plt.streamplot(x_grid, y_grid, Bx, By)

# Show the plot
plt.show()
```

This is a code which illustrates the magnetic field distribution when a current carrying loop is placed in Z = 0 and then plot is taken in a plane parallel to it i.e. Z != 0. For reference and better visualisation I have plotted this in Z = 1 keeping the loop at Z = 0.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid of points to evaluate the electric field
x_grid, y_grid = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))

dtheta = (2*np.pi)/100
R = 2
I = 5
Z = 1
k = 10**(-7)*R*I

Bx, By = 0, 0

for theta in np.linspace(0, 2*np.pi, 100):
    Bx += (Z*R*dtheta*np.cos(theta))/((x_grid+R*np.cos(theta))**2 + (y_grid-R*np.sin(theta))**2 + Z**2)**(1.5)
    By += (-Z*R*dtheta*np.sin(theta))/((x_grid+R*np.cos(theta))**2 + (y_grid-R*np.sin(theta))**2 + Z**2)**(1.5)

# Plot the electric field lines
plt.streamplot(x_grid, y_grid, Bx, By)

# Show the plot
plt.show()
```
